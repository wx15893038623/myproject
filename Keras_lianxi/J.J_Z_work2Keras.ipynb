```python
import numpy as np
import pandas as pd
import keras
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
dataframe = pd.read_csv('winequality-red.csv')
```

```python
dataframe.head()
```

```python
dataset = dataframe.values
```

```python
x = dataset[:,0:11].astype(float)
```

```python
x[1]
```

```python
y1 = dataset[:,11]
```

```python
y1[1]
```

```python
#质量按非零整数标号
encoder = LabelEncoder()
encoded_Y = encoder.fit_transform(y1)
```

```python
#将整数转换为虚拟变量(one hot encoding)
y = np_utils.to_categorical(encoded_Y)
```

```python
y[1]
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(output_dim=2000, input_dim=11, activation='relu'))
    model.add(Dense(output_dim=6, input_dim=2000, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='RMSProp', metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=500, batch_size=40)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
estimator.fit(X_train, Y_train)
```

```python
pred = estimator.predict(X_test)
```

```python
estimator.score(X_test,Y_test)
```

```python
seed = 42
np.random.seed(seed)
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, x, y, cv=kfold)
```
