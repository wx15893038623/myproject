```python
24 ms	15.5 MB
class Solution(object):
    def splitListToParts(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: List[ListNode]
        """
        cnt = 0 
        ans = []
        cur = head
        while cur:
            cnt += 1
            cur = cur.next
        if cnt <= k:
            while head:
                temp = head.next
                head.next = None
                ans.append(head)
                head = temp
            return ans+[None]*(k-cnt)
        else:
            part_len = cnt // k
            residual = cnt % k
            j = 1
            while head:
                cur,temp = head,head
                count = 0
                if j <= residual:
                    while count < part_len + 1:
                        count += 1
                        cur = temp
                        temp = temp.next
                    cur.next = None
                    j += 1
                    ans.append(head)
                else:
                    while count < part_len:
                        count += 1
                        cur = temp
                        temp = temp.next
                    cur.next = None
                    ans.append(head)
                head = temp
            return ans
```
